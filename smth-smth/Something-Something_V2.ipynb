### Something-Something dataset : 
사람이 일상적인 기본 동작을 수행하는 것을 보여주는 비디오 모음
- 각 비디오에 레이블이 달려있음(ex)테이블에 컵을 두다)
- 객체 주석이 추가로 달려있음(ex)어떤 것에 어떤 것을 두다)
- 비디오의 형식은 webm이다.
- 174class로 나누어진 비디오 데이터셋
- 240px, 12frames/sec
